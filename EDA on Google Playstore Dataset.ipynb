```python
data_copy['Reviews']=data_copy['Reviews'].astype(int)
```

```python
data_copy['Size'].unique()
```

```python
def convert_size(value):
    if isinstance(value, str):
        if 'M' in value: 
            return float(value.replace('M', ' ')) * 1000
        elif 'k' in value:  
            return float(value.replace('k', ' ')) * 1
    return value

data_copy['Size']=data_copy['Size'].apply(convert_size)
data_copy['Size'].unique()
```

```python
data_copy['Size'] = data_copy['Size'].replace("Varies with device", np.nan)
data_copy['Size'] = pd.to_numeric(data_copy['Size'])
data_copy['Size'] = data_copy['Size'].fillna(0).astype(int)
```

```python
data_copy['Size'].unique()
```

```python
data_copy.head(2)
```

```python
data_copy=data_copy.rename(columns={'Size':'Size(kb)'})
```

```python
data_copy.head()
```

```python
data_copy['Installs'].unique()
```

```python
data_copy['Type'].unique()
```

```python
data_copy['Type'].dropna()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']

for chars in chars_to_remove:
    for cols in cols_to_clean:
        data_copy[cols]=data_copy[cols].str.replace(chars,'')
```

```python
data_copy['Installs'].unique()
```

```python
data_copy['Installs']=data_copy['Installs'].astype(int)
```

```python
data_copy['Price'].unique()
```

```python
data_copy['Price']=data_copy['Price'].astype(float)
```

```python
data_copy.head()
```

```python
data_copy['Last Updated']=pd.to_datetime(data_copy['Last Updated'])
data_copy['Day']=data_copy['Last Updated'].dt.day
data_copy['Month']=data_copy['Last Updated'].dt.month
data_copy['Year']=data_copy['Last Updated'].dt.year
```

### FEATURE INFORMATION

1. **APP**: Name of the application  
2. **CATEGORY**: Category under which the application falls  
3. **RATING**: Rating of application at Play Store  
4. **REVIEWS**: Number of reviews of application on Play Store  
5. **SIZE**: Size of application  
6. **INSTALLS**: Number of installs of the app  
7. **TYPE**: Free or paid app  
8. **PRICE**: Price of the app  
9. **CONTENT RATING**: Target audience category  
10. **GENRE**: Genre to which the app belongs  
11. **CURRENT VER**: Current version of the application available  
12. **ANDROID VER**: Android version of application  
13. **DAY**: Day of latest update  
14. **MONTH**: Month of latest update  
15. **YEAR**: Year of latest update  


```python
## prportion of count data on categorical column
for cols in categorical_features:
    print(data_copy[cols].value_counts(normalize=True)*100)
    print("---------------------")
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=data_copy[numeric_features[i]],shade=True,color='r') ##kernel density estimation
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

OBSERVATION: 'Rating' and 'Year' are left skewed while 'Reviews','Size','Installs' and 'Price' are right skewed.

```python
plt.figure(figsize=(15,10))
plt.suptitle('Univariate Analysis of Categorical Features',fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category= ['Type','Content Rating']

for i in range(0, len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=data_copy[category[i]],palette='hls')
    plt.xlabel(category[i])
    plt.xticks(rotation=0)
    plt.tight_layout()
```

OBSERVATION: Free apps are the dominating ones.
